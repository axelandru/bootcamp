````python
def f2(n):
    for i in range(1, n):
        print(f"~~~ {i}")
        newdict1 = dict({'cubed':i**3})
        newdict2 = dict({'squared':i**2})
        yield newdict1
        yield newdict2
        print(f"@@@{i}")




n = 5
for x in f2(n):
    print(x)
````
